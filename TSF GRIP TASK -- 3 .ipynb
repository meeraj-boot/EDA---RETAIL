```python
# areas where i can make money
df=pd.read_csv("sample.csv")
```

```python
df
```

```python
#calculate total sales and profit
category_sales = df.groupby('Category')['Sales'].sum().reset_index()
```

```python
category_profit = df.groupby('Category')['Profit'].sum().reset_index()
```

```python
category_profit 
```

```python
category_sales
```

```python
#merge
category_data = pd.merge(category_sales, category_profit, on='Category')
category_data
```

```python
#profit margin in category
category_data['Profit Margin'] = (category_data['Profit'] / category_data['Sales']) * 100
category_data['Profit Margin'] 
```

```python
#visualization of profit
#bar
sns.barplot(x='Category', y='Profit Margin', data=category_data)
plt.title('Profit Margin by Category')
plt.xlabel('Category')
plt.ylabel('Profit Margin')
plt.show()
```
